```python
import numpy as np
import matplotlib.pyplot as plt

A = np.array([[[0.,0.],[0.,0.25]],
              [[0.95,0.005],[-0.005,0.85]],
              [[0.35,-0.2],[0.16,0.04]],
              [[-0.04,0.2],[0.16,0.04]]])
    
B = np.array([[0.,-0.4],
              [-0.002,0.5],
              [-0.09,0.02],
              [0.083,0.12]])

ntrials = 40000
pos = np.zeros((ntrials,2))
for n in range(1,ntrials):
    type = np.random.choice(range(4),p=[0.02,0.84,0.07,0.07])
    pos[n] = A[type].dot(pos[n-1])+B[type]

plt.figure(figsize=(8, 12), dpi=80)
plt.scatter(pos[:,0],pos[:,1],c=pos[:,1]*0.5,alpha=0.5,s=10,marker='.')
plt.show()
```
